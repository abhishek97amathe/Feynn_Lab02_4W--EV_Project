```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ev_cat.csv')
df.tail()  # Shows the first 5 rows for a quick preview
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month
```

```python
# check the info of the dataframe
df.info()
```

```python
# 5 point summary of the dataframe
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.sample(5)  # Shows a random sample of 5 rows to check the data
```

```python
# find cumsum of each vehicle category per year
vehicle_cols = [col for col in df.columns if col not in ['Date', 'Year', 'Month']]
for col in vehicle_cols:
	df[f'{col}_cumsum'] = df.groupby('Year')[col].cumsum()

```

```python
# find cumsum of each cat per month 
for col in vehicle_cols:
    df[f'{col}_cumsum_month'] = df.groupby(['Year', 'Month'])[col].cumsum()
```

```python
#find the valuecount of each vehicle category per year
vehicle_counts = df.groupby('Year')[vehicle_cols].sum().reset_index()
vehicle_counts
```

```python
#find the valuecount of each vehicle category per month
vehicle_counts_month = df.groupby(['Year', 'Month'])[vehicle_cols].sum().reset_index()
```

```python
vehicle_counts_month
```

```python
#PLOT THE  BAR PLOT OF EACH VEHICLE CATEGORY PER YEAR
plt.figure(figsize=(12, 6)) 
vehicle_counts.plot(x='Year', kind='bar', stacked=True, figsize=(12, 6))
plt.title('Electric Vehicle Categories per Year')   
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.legend(title='Vehicle Categories')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
#PLOT THE  BAR PLOT OF TOTAL VEHICLES PER YEAR
plt.figure(figsize=(12, 6))
vehicle_counts['Total'] = vehicle_counts[vehicle_cols].sum(axis=1)  
vehicle_counts.plot(x='Year', y='Total', kind='bar', color='skyblue', figsize=(12, 6))
plt.title('Total Electric Vehicles per Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Vehicles')  
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
#CREATE A TABLE TOTAL VEHICLES PER YEAR
total_vehicles_per_year = vehicle_counts[['Year', 'Total']]
total_vehicles_per_year
```

```python
#PLOT THE  LINE PLOT OF TOTAL VEHICLES PER QUARTER
vehicle_counts['Quarter'] = pd.to_datetime(vehicle_counts['Year'].astype(str) + '-01-01').dt.to_period('Q') 
plt.figure(figsize=(12, 6))
vehicle_counts_quarter = vehicle_counts.groupby('Quarter').sum().reset_index()
vehicle_counts_quarter.plot(x='Quarter', y='Total', kind='line', marker='o', figsize=(12, 6))
plt.title('Total Electric Vehicles per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Number of Vehicles')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
